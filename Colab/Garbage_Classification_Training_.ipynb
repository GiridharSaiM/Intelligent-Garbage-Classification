Import Image Data Generator Library

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

Setting parameter for image augmentation to the training data

```python
train_datagen=ImageDataGenerator(rescale=1./255,shear_range=0.1,zoom_range=0.1,horizontal_flip=True)
```

Image data augmentation to the testing data

```python
val_datagen=ImageDataGenerator(rescale=1./255)
```

Loading our data and performing data augmentation

```python
train_transform=train_datagen.flow_from_directory("/content/drive/MyDrive/Garbage3/Dataset/Train",target_size=(128,128),batch_size=64,class_mode="categorical")
```

```python
test_transform=val_datagen.flow_from_directory("/content/drive/MyDrive/Garbage3/Dataset/Test",target_size=(128,128),batch_size=64,class_mode="categorical")
```

Importing Libraries

```python
#Import Sequential Model
from tensorflow.keras.models import Sequential
#Import Dense Layer
from tensorflow.keras.layers import Dense
#Import Convolution2D Layer
from tensorflow.keras.layers import Convolution2D
#import MaxPooling2d Layer
from tensorflow.keras.layers import MaxPooling2D
#Adding Flatten Layer
from tensorflow .keras. layers import Flatten
from tensorflow .keras .optimizers import Adam
```

Intializing The Model

```python
model=Sequential()
```

First Convolution Layer and pooling

```python
model.add(Convolution2D(32,(3,3),input_shape=(128,128,3),activation="relu"))
model.add(MaxPooling2D(2,2))
```

Second Convolution Layer and pooling

```python
model.add(Convolution2D(64,(3,3),padding="same",activation="relu"))
```

input shape is going to be pooled feature maps for the previous convolution

```python
model.add(MaxPooling2D(2,2))
```

Third Convolution layer and pooling

```python
model.add(Convolution2D(32,(3,3),activation="relu"))
model.add(MaxPooling2D(2,2))
```

Fourth Convolution layer and pooling


```python
model.add(Convolution2D(32,(3,3),padding="same",activation='relu'))
```

input shape is going to be pooled feature maps for the previous convolution

```python
model.add(Convolution2D(32,(3,3),padding="same",activation='relu'))
```

Flattening The Layers

```python
model.add(Flatten())
```

Adding Fully Connected Layers

```python
#Adding First Hidden Layer
model.add(Dense(kernel_initializer="uniform",activation="relu",units=150))
#Adding Second Hidden Layer
model.add(Dense(kernel_initializer="uniform",activation="relu",units=68))
model.add(Dense(kernel_initializer="uniform",activation="softmax",units=6))
```

Summary Of The Model

```python
model.summary()
```

Compiling The CNN Model

```python
model.compile(loss="categorical_crossentropy",optimizer="adam",metrics=["acc"])
```

Fit The Model

```python
result=model.fit(train_transform,steps_per_epoch=2527/64,validation_steps=526/64,epochs=30,validation_data=test_transform)
```

Save The Model

```python
model.save("Garbage3.h5")
```
